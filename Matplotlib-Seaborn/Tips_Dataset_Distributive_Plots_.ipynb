```python
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
```

```python

# seaborn library has some datasets , we will use tips datasets ...
```

```python
df = sns.load_dataset('tips')
display(df)
df.info()
display(df.head())
```

```python
#-------------TO VISUALIZE QUANTATIVE DATA OR COLUMNS WE USE DISTRIBUTIVE PLOTS --------------------#

# HISTOGRAMS ARE BEST TO VISUALIZE DISTRIBUTIVE PLOTS
```

```python
# --------------------TO VISUALIZE ONE QUANTATIVE DATA BY HISTOGRAM --> HISTPLOT -----------------------#

plt.subplot(1,2,1)
print(sns.histplot(df['total_bill'],bins=10,kde=True))

plt.subplot(1,2,2)
print(sns.histplot(df['tip'],bins=10,kde=True))
# we can see the distribution of curve and its skewness right or left skewness
# kde is kernel density estimator --> for skewness it plots variance 

#-----------------KDE-----------------------------#
'''
KDE is kernel density estimator . It calcualtes probability density for every x 

Algorithm takes an excat spot or example X=25 is taken 
It calculates how far are every data point compared to X=25
Farther points give lower score and Nearer gives higher and sums all scores 
Then it moves to tiny fraction right ( 25.01 ) and repeats the same process 

How it decides the neighbour window which is called as Bandwidth? --> By variance 
'''
```

```python
#-------------TO VISUALIZE TWO QUANTATIVE DATA AT SAME TIME  ----> JOINTPLOT -----------------#


print(sns.jointplot(x='total_bill' , y='tip' , data =df, kind ='scatter'))


print(sns.jointplot(x='total_bill',y='tip',data=df , kind ='hex'))

print(sns.jointplot(x='total_bill' , y= 'tip' , data =df , kind = 'reg'))
```

```python
#-------------------------TO VISUALIZE ALL COMBINATIONS WITH ALL PLOTS IN A SINGLE CANVAS --> PAIR PLOT -------------------------------------------#

sns.pairplot(df,hue='sex', kind ='scatter')

# to visualize data with some parameters or degress like i want to male and female contribution in a scatter plot ---> hue

#---------hue----------#

# hue is used to colour code my data points w.r.t specified unique values in categorical columns like male and female as diff colours
# column in hue to colour data points must be categorical ( qualitative not quantative )
# categorical column must need fewer unique values so not much data points with unique colours as messy
```
