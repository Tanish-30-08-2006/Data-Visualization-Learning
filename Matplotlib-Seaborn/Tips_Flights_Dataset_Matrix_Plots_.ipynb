```python
import numpy as np 
from IPython.display import display
import seaborn as sns
import pandas as pd
```

```python
flights = sns.load_dataset('flights')
display(flights)

tips = sns.load_dataset('tips')
display(tips)
```

```python
# For correlation we only track numerical values 
tips_corr = tips[['total_bill','tip','size']]
display(tips_corr)
```

```python
display(tips_corr.corr()) # correlation is a method which must be called with () bracket
# >0 -> positive and <0 negative   
#[-1,+1] goes values of correlation
```

```python
# For correlation we only track numerical values 
tips_corr = tips[['total_bill','tip','size']]
display(tips_corr)
```

```python
sns.heatmap(tips_corr.corr(),annot=True)
# Annot puts values in boxes  
display(tips_corr.corr())
```

```python
#--------------CLUSTER MAP --------------# similar to heat map 
sns.clustermap(tips_corr.corr(),annot=True)

# Heat map of a dataset has a fixed numerical or alphabeitcal order of columns of the dataset
# Clustermap shuffles the order and pushed similar data patterns right next to each other ...
```
