```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import display
```

```python
#-----------------------TO VISUALIZE QUALITATIVE DATA OR VARIABLES(COLUMNS) WE USE CATEGORICAL PLOTS---------------#

# here sex smoker day time and size are quantative variables here 


# generally on y axis is quantative variable ( count or anyone ) 
#           on x axis is qualitative variable ( gender smoker size)
```

```python
df = sns.load_dataset("tips")
display(df)
```

```python
#--------------COUNT PLOTS---------------#

display(df.head())
sns.countplot(x=df['sex'] , hue = df['smoker'],data=df)
plt.show()

sns.countplot(x=df['smoker'])
plt.show()

sns.countplot(x=df['sex'])
plt.show()
```

```python
#-----------------------BARPLOT--------------------#

# By default , comparison type is mean of qantitative data but we x=can change it ...

sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.max) 
#remember method vs attributes shape and size are attributes of array and min max are methods 
# here we need to just pass the function not the result of it np.max() is a number / result as a return value , we need function 
plt.show()

sns.barplot(x=df['sex'] , y= df['total_bill'],estimator = np.mean)
plt.show()

```

```python
#-----------------BOX PLOTS AND VIOLIN PLOTS --------------#

# Box plot --> used to show distribution of cateforical data 
               # used to detect outliers 
               # centre line shows the median value of data we used and 
               # and top and bottom are 25 AND 75th percentile data and box contains 50% of data 
               # IQR - Interquartile range which is Height of box and Boundaries are set as 1.5*IQR 
               # ANY value outside this boundary is drawn as data point which shows an outlier .......

sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
#-----------------------------VIOLIN PLOT---------------------------------#

# Violin plot  --> combination of box plot and KDE , wider curve is more data points at that value and extremeties define the standard deviation..

sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow')
plt.show()

sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
plt.show()
```

```python
#-------------------------------SWARM and STRIP PLOT------------------------------#

# Similar to Violin plot but shows data points instead of a distribution 
# STRIP PLOT --> shows a strip not a curve or a spread

sns.swarmplot(x='total_bill',y='day',data=df)
plt.show()
sns.stripplot(x='total_bill',y='day',data=df)
plt.show()
```

```python
# SWARM AND VIOLIN PLOT IN SAME CANVAS 
sns.violinplot(x='total_bill',y='day',data=df)
sns.swarmplot(x='total_bill',y='day',data =df )
```
